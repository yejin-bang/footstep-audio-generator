What I should see - folder structure


Project Name/
├── ai_env/           (your virtual environment - lots of files)
├── data/             (empty folder)
├── models/           (empty folder)
├── src/              (empty folder)
├── notebooks/        (empty folder)
├── tests/            (empty folder)
├── README.md         (empty file)
├── requirements.txt  (list of your packages)
└── .gitignore       (empty file)

In Terminal:

mkdir data models src notebooks tests
touch README.md requirements.txt .gitignore

Save my package list:
pip freeze > requirements.txt

In VSCode terminal: 

Activate my virtual environment
source project_name/bin/activate
pip install what-you-need
