```python
from diffusers import AudioLDMPipeline
import torch
import scipy.io.wavfile
from IPython.display import Audio, display
```

```python
pipe = AudioLDMPipeline.from_pretrained("cvssp/audioldm-s-full-v2")
pipe = pipe.to("cpu")

print("AudioLDM loaded and ready!")
print("We can now turn text into audio!")
```

```python
print("Creating out AI-generated sound...")

prompt = "High quality,clear audio, drum beats, 4/4"

print(f'Generating: {prompt}')

audio = pipe(
    prompt,
    num_inference_steps=50,
    audio_length_in_s=8.0,
    guidance_scale=7.5,
    num_waveforms_per_prompt=1
).audios[0]

print("Audio generated!")
print(f'Audio shape: {audio.shape}')
print(f'Audio is an array of {len(audio)} numbers representing sound waves')



print("Playing your AI-generated rain sound:")
display(Audio(audio, rate=16000))
```
